```python
# Import pandas and matplotlib first
import pandas as pd
import matplotlib.pyplot as plt

# Create or load your DataFrame
# Option 1: Create a sample DataFrame
df = pd.DataFrame({
    'restaurant': ['Restaurant A', 'Restaurant B', 'Restaurant A', 'Restaurant C', 'Restaurant B'],
    'calories': [500, 700, 450, 800, 650]
})

# Option 2: Load from a file (uncomment and modify as needed)
# df = pd.read_csv('your_file.csv')

# Calculate average calories by restaurant
avg_calories = df.groupby('restaurant')['calories'].mean()

# Plot the results
avg_calories.plot(kind='bar')
plt.title("Average Calories per Restaurant")
plt.show()
```

```python
# Import the required libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd  # Added pandas import

# Create or load your DataFrame
# Option 1: Load data from a file
# df = pd.read_csv('your_data_file.csv')  # Uncomment and modify this line to load your data

# Option 2: For demonstration, create a sample DataFrame
df = pd.DataFrame({
    'calories': [150, 200, 300, 250, 180],
    'protein': [5, 10, 15, 12, 8],
    'sodium': [200, 300, 400, 250, 350],
    'total_fat': [6, 12, 18, 15, 9]
})

numeric_cols = ['calories', 'protein', 'sodium', 'total_fat']

# Create the heatmap
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')

plt.title("Correlation Heatmap")
plt.show()
```
